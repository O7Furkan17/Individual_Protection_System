```python
import os
import random
import shutil
import json
import cv2
```

```python
#json dosyalarını txt çevir
def convert_all_json_to_txt(json_dir, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(json_dir):
        if filename.endswith('.json'):
            json_file = os.path.join(json_dir, filename)
            with open(json_file, 'r') as f:
                data = json.load(f)

            for obj in data['objects']:
                if obj['classTitle'] == 'knife':
                    xmin, ymin = obj['points']['exterior'][0]
                    xmax, ymax = obj['points']['exterior'][1]
                    txt_format = f"knife {xmin} {ymin} {xmax} {ymax}"

                    txt_filename = os.path.splitext(os.path.splitext(filename)[0])[0] + '.txt'
                    output_file = os.path.join(output_dir, txt_filename)
                    with open(output_file, 'a') as out_f:
                        out_f.write(txt_format + '\n')

json_directory = '/content/drive/MyDrive/ds/ann'
output_directory = '/content/output'
convert_all_json_to_txt(json_directory, output_directory)
```

```python
#txt içinde yazan class isimlerini düzenle
def replace_class_in_txt_files(txt_dir, new_class_id):
    for filename in os.listdir(txt_dir):
        if filename.endswith('.txt'):
            txt_file = os.path.join(txt_dir, filename)
            with open(txt_file, 'r') as file:
                lines = file.readlines()

            with open(txt_file, 'w') as file:
                for line in lines:
                    parts = line.strip().split()
                    parts[0] = str(new_class_id)
                    file.write(" ".join(parts) + '\n')

txt_directory = '/content/drive/MyDrive/Knife/labels'
new_class_id = 1
replace_class_in_txt_files(txt_directory, new_class_id)
```

```python
#dosyaları başka bir yere kopyalama
def copy_all_files(src_dir, dest_dir):
    i=0
    for filename in os.listdir(src_dir):
        src_file = os.path.join(src_dir, filename)
        dest_file = os.path.join(dest_dir, filename)

        if os.path.isfile(src_file):
            shutil.copy(src_file, dest_file)
            print(f"Copied: {src_file} to {dest_file}")
            i = i + 1
    print("Toplam dosya sayısı: ",i)

source_directory = '/content/drive/MyDrive/Gun/labels'
destination_directory = '/content/drive/MyDrive/dataset/labels/train'
copy_all_files(source_directory, destination_directory)
```

```python
#görselleri normalize etme
def normalize_coordinates(image_path, label_path):
    image = cv2.imread(image_path)
    h, w, _ = image.shape

    with open(label_path, 'r') as file:
        lines = file.readlines()

    normalized_lines = []
    for line in lines:
        class_id, x_center, y_center, width, height = map(float, line.strip().split())
        if x_center >1 or y_center >1 or width > 1 or height >1:
            x_center_norm = x_center / w
            y_center_norm = y_center / h
            width_norm = width / w
            height_norm = height / h
            normalized_lines.append(f"{int(class_id)} {x_center_norm} {y_center_norm} {width_norm} {height_norm}\n")
            print("Normalize edildi:", label_path)
        else:
            normalized_lines.append(line)

    with open(label_path, 'w') as file:
        file.writelines(normalized_lines)

directory = "/content/drive/MyDrive/Knife/images"

for root, _, files in os.walk(directory):
    for file in files:
        if file.endswith(".jpg"):
            image_path = os.path.join(root, file)
            label_path = os.path.join(root.replace("images", "labels"), file.replace(".jpg", ".txt"))
            normalize_coordinates(image_path, label_path)
        elif file.endswith(".JPG"):
            image_path = os.path.join(root, file)
            label_path = os.path.join(root.replace("images", "labels"), file.replace(".JPG", ".txt"))
            normalize_coordinates(image_path, label_path)
```

```python
#val dosyalarını seçme
def move_images_and_labels(num_images_to_move):
    dataset_dir = "/content/drive/MyDrive/dataset"

    images_dir = os.path.join(dataset_dir, "images")

    train_dir = os.path.join(images_dir, "train")
    val_dir = os.path.join(images_dir, "val")

    train_images = [f for f in os.listdir(train_dir) if f.endswith(".jpg")]

    images_to_move = random.sample(train_images, num_images_to_move)

    for image_name in images_to_move:
        src_image = os.path.join(train_dir, image_name)
        dst_image = os.path.join(val_dir, image_name)
        shutil.move(src_image, dst_image)

        parts = image_name.rsplit(".jpg", 1)
        label_name = parts[0] + ".txt"
        src_label = os.path.join(dataset_dir, "labels", "train", label_name)
        dst_label = os.path.join(dataset_dir, "labels", "val", label_name)
        shutil.move(src_label, dst_label)

move_images_and_labels(150)
```

```python
#txt içindeki verileri kontrol etme
def check_txt_files_in_colab_directory(directory):
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            file_path = os.path.join(directory, filename)
            if os.path.isfile(file_path):
                with open(file_path, 'r', encoding='utf-8') as file:
                    content = file.read().strip()
                    if not content or content == ' ' or content == '':
                        print(f'{filename} is empty.')

directory = '/content/drive/MyDrive/dataset/labels/val'
check_txt_files_in_colab_directory(directory)
```

```python
#boş dosyaları silme
def delete_empty_txt_and_corresponding_images_in_colab(labels_directory, images_directory):
    for filename in os.listdir(labels_directory):
        if filename.endswith('.txt'):
            txt_file_path = os.path.join(labels_directory, filename)
            image_file_path = os.path.join(images_directory, filename.replace('.txt', '.jpg'))

            if os.path.isfile(txt_file_path):
                with open(txt_file_path, 'r', encoding='utf-8') as file:
                    content = file.read().strip()
                    if not content or content == ' ' or content == '':
                        try:
                            os.remove(txt_file_path)
                            print(f'Removed empty .txt file: {txt_file_path}')
                        except Exception as e:
                            print(f'Failed to delete {txt_file_path}. Reason: {e}')

                        if os.path.isfile(image_file_path):
                            try:
                                os.remove(image_file_path)
                                print(f'Removed corresponding image file: {image_file_path}')
                            except Exception as e:
                                print(f'Failed to delete {image_file_path}. Reason: {e}')

labels_directory = '/content/drive/MyDrive/Knife/labels'
images_directory = '/content/drive/MyDrive/Knife/images'

delete_empty_txt_and_corresponding_images_in_colab(labels_directory, images_directory)
```
